Matplotlib Tutorial (Matplotlib Eğitimi)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# grafiğin x eksenleri sırasıyla (1,2,6,8) şeklindedir bu noktaların y ekseni ise alta bulunan (3,8,1,10) değerleridir 
# örneğin x eksenlerini belirtmemiş olalım, python bunu varsayılan olarak ayarlar bunuda y noktalarının değer uzunluğu kadar 0'dan başlayarak ayarlar.
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

#plot fonksiyonuna önce x sonra y kordinatlarını giriyoruz.
plt.plot(xpoints, ypoints)
# show fonksiyonu ile grafiği yazdırıyoruz.
plt.show()
```

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])

# plot fonksiyonunun içine "o" konulunca belirtilen kordinatlara nokta koyar.
plt.plot(xpoints,ypoints,"o")
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])


plt.plot(xpoints,ypoints, marker = 'o', ms = 20, mec = 'b', mfc = 'r')
plt.show()
```

```python
# Çizgi stili daha kısa bir söz dizimi ile yazılabilir:
# linestyle, ls olarak yazılabilir.
# dotted, ":" olarak yazılabilir.
# dashed,"--" olarak yazılabilir.
# dashdot, '-.'olarak yazılabilir.

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

# ls yazı tipini belirtiyor, color yazının rengini ayarlıyoruz, linewidth ile çizginin kalınlığını belirtiyoruz.
plt.plot(xpoints, ypoints,ls = ':',color = 'r',linewidth = '4.5')
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# yazı tipi rengi ve yazının büyüklüğünü aarlıyoruz ve font1 olarak yazdırıyoruz.
font1 = {'family':'serif','color':'blue','size':20}
# yazı tipi rengi ve yazının büyüklüğünü aarlıyoruz ve font2 olarak yazdırıyoruz.
font2 = {'family':'serif','color':'darkred','size':15}

# title fonksiyonu ile grafiğimize başlık ekliyoruz parametreler (başlık,yazı fontu)
plt.title("Sports Watch Data", fontdict = font1)

# xlable fonksiyonu ile grafiğimize x eksenine başlık ekliyoruz parametreler (başlık,yazı fontu)
plt.xlabel("Average Pulse", fontdict = font2)

# ylable fonksiyonu ile grafiğimize y eksenine başlık ekliyoruz parametreler (başlık,yazı fontu)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.grid()
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

# Grafiğin arkasına çizgi çekmemizi sağlıyor ve grafikte kullandığımız özelikler bu çizgilerde de geçeridir.(color,linestyle,inewidth)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

# grafiğin arkasına çizilen çizgileri y eksenine dik şekilde çizer.
# plt.grid(axis = 'y')

# grafiğin arkasına çizilen çizgileri x eksenine dik şekilde çizer.
plt.grid(axis = 'x')

plt.show()
```

```python
# birden fazla grafik çizmek için subplot fonksiyonu kullanılır. subplot fonksiyonu üç argüman alır (satır,sütün,grafiğin konumu)

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1'nci satır, 2'nci sütun,grafiğin konumu olan 1 yazılır.
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1'nci satır, 2'nci sütun,grafiğin konumu olan 2 yazılır.
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
# noktaların x ve y kordinatları. 
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# scatter fonksiyonu ile x ve y  noktalarını çizdirebiliyoruz.
plt.scatter(x, y)

# noktaların x ve y kordinatları. 
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

# scatter fonksiyonu ile x ve y  noktalarını çizdirebiliyoruz.
plt.scatter(x, y)

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# her noktanın renklerini renk haritasının  aralığına göre bir değer vererek renklendirdik.
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# yerleşik renk haritası olan cmap değişkenine renk haritasını kullanacağımızdan "viridis" anahtar kelimesini atıyoruz.
plt.scatter(x, y, c=colors, cmap='viridis')

# color bar fonksiyonu grafiğin yanında renk haritasını gösterilmesini sağlar.
plt.colorbar()

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

#noktaların boyut dizesi.
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

# noktaların boyutlarını s argümanı ile değiştirebiiyoruz bun belirlediğimiz sizes değişkenini atıyoruz.
# alpha argümanı ile noktaları şeffaf hale getirebiiyoruz.
plt.scatter(x, y, s=sizes,alpha=0.5)

plt.show()
```

```python
# sutunların x değereri.
x = np.array(["A", "B", "C", "D"])
# sutunların y değereri.
y = np.array([3, 8, 1, 10])

# bar fonksiyonu ile sutun grafiklerini çizdirebiliyoruz parametreler (x,y,rengi,sutunların kalınlıkları).
plt.bar(x,y,color = "red",width = 0.1)

plt.show()
```

```python
# sutunların x değereri.
x = np.array(["A", "B", "C", "D"])
# sutunların y değereri.
y = np.array([3, 8, 1, 10])

# bar fonksiyonu ile sutun grafiklerini çizdirebiliyoruz parametreler (x,y,rengi,sutunların kalınlıkları).
plt.barh(x, y,color = "red",height = 0.1)

plt.show()
```

```python
# pasta grafiğinde x ekseninin pozitif yönünden başlar ve saat yönünün tersinde haraket eder. 
y = np.array([35, 25, 25, 15])

# parçaları isimlendirmek için isimleri bir liste içine alıyoruz.
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# parçalar arasındaki boşlukları ifade eden değerleri listeliyoruz.
myexplode = [0.2, 0, 0, 0]

# lables parametresine isimleriizin listesinin bulunduğu mylables değişkenini atıyoruz, startangle parametresi ile grafiğimizi döndürebiliyoruz suan saat yönünün tersine 90 derece döndürdük.
plt.pie(y,labels = mylabels, startangle = 90, explode = myexplode,shadow = True)
plt.show() 
```
