Matplotlib Tutorial (Matplotlib EÄŸitimi)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# grafiÄŸin x eksenleri sÄ±rasÄ±yla (1,2,6,8) ÅŸeklindedir bu noktalarÄ±n y ekseni ise alta bulunan (3,8,1,10) deÄŸerleridir 
# Ã¶rneÄŸin x eksenlerini belirtmemiÅŸ olalÄ±m, python bunu varsayÄ±lan olarak ayarlar bunuda y noktalarÄ±nÄ±n deÄŸer uzunluÄŸu kadar 0'dan baÅŸlayarak ayarlar.
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

#plot fonksiyonuna Ã¶nce x sonra y kordinatlarÄ±nÄ± giriyoruz.
plt.plot(xpoints, ypoints)
# show fonksiyonu ile grafiÄŸi yazdÄ±rÄ±yoruz.
plt.show()
```

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])

# plot fonksiyonunun iÃ§ine "o" konulunca belirtilen kordinatlara nokta koyar.
plt.plot(xpoints,ypoints,"o")
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])


plt.plot(xpoints,ypoints, marker = 'o', ms = 20, mec = 'b', mfc = 'r')
plt.show()
```

```python
# Ã‡izgi stili daha kÄ±sa bir sÃ¶z dizimi ile yazÄ±labilir:
# linestyle, ls olarak yazÄ±labilir.
# dotted, ":" olarak yazÄ±labilir.
# dashed,"--" olarak yazÄ±labilir.
# dashdot, '-.'olarak yazÄ±labilir.

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

# ls yazÄ± tipini belirtiyor, color yazÄ±nÄ±n rengini ayarlÄ±yoruz, linewidth ile Ã§izginin kalÄ±nlÄ±ÄŸÄ±nÄ± belirtiyoruz.
plt.plot(xpoints, ypoints,ls = ':',color = 'r',linewidth = '4.5')
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# yazÄ± tipi rengi ve yazÄ±nÄ±n bÃ¼yÃ¼klÃ¼ÄŸÃ¼nÃ¼ aarlÄ±yoruz ve font1 olarak yazdÄ±rÄ±yoruz.
font1 = {'family':'serif','color':'blue','size':20}
# yazÄ± tipi rengi ve yazÄ±nÄ±n bÃ¼yÃ¼klÃ¼ÄŸÃ¼nÃ¼ aarlÄ±yoruz ve font2 olarak yazdÄ±rÄ±yoruz.
font2 = {'family':'serif','color':'darkred','size':15}

# title fonksiyonu ile grafiÄŸimize baÅŸlÄ±k ekliyoruz parametreler (baÅŸlÄ±k,yazÄ± fontu)
plt.title("Sports Watch Data", fontdict = font1)

# xlable fonksiyonu ile grafiÄŸimize x eksenine baÅŸlÄ±k ekliyoruz parametreler (baÅŸlÄ±k,yazÄ± fontu)
plt.xlabel("Average Pulse", fontdict = font2)

# ylable fonksiyonu ile grafiÄŸimize y eksenine baÅŸlÄ±k ekliyoruz parametreler (baÅŸlÄ±k,yazÄ± fontu)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.grid()
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

# GrafiÄŸin arkasÄ±na Ã§izgi Ã§ekmemizi saÄŸlÄ±yor ve grafikte kullandÄ±ÄŸÄ±mÄ±z Ã¶zelikler bu Ã§izgilerde de geÃ§eridir.(color,linestyle,inewidth)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

# grafiÄŸin arkasÄ±na Ã§izilen Ã§izgileri y eksenine dik ÅŸekilde Ã§izer.
# plt.grid(axis = 'y')

# grafiÄŸin arkasÄ±na Ã§izilen Ã§izgileri x eksenine dik ÅŸekilde Ã§izer.
plt.grid(axis = 'x')

plt.show()
```

```python
# birden fazla grafik Ã§izmek iÃ§in subplot fonksiyonu kullanÄ±lÄ±r. subplot fonksiyonu Ã¼Ã§ argÃ¼man alÄ±r (satÄ±r,sÃ¼tÃ¼n,grafiÄŸin konumu)

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1'nci satÄ±r, 2'nci sÃ¼tun,grafiÄŸin konumu olan 1 yazÄ±lÄ±r.
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1'nci satÄ±r, 2'nci sÃ¼tun,grafiÄŸin konumu olan 2 yazÄ±lÄ±r.
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# her noktanÄ±n renklerini renk haritasÄ±nÄ±n  aralÄ±ÄŸÄ±na gÃ¶re bir deÄŸer vererek renklendirdik.
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

# yerleÅŸik renk haritasÄ± olan cmap deÄŸiÅŸkenine renk haritasÄ±nÄ± kullanacaÄŸÄ±mÄ±zdan "viridis" anahtar kelimesini atÄ±yoruz.
plt.scatter(x, y, c=colors, cmap='viridis')

# color bar fonksiyonu grafiÄŸin yanÄ±nda renk haritasÄ±nÄ± gÃ¶sterilmesini saÄŸlar.
plt.colorbar()

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

#noktalarÄ±n boyut dizesi.
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

# noktalarÄ±n boyutlarÄ±nÄ± s argÃ¼manÄ± ile deÄŸiÅŸtirebiiyoruz bun belirlediÄŸimiz sizes deÄŸiÅŸkenini atÄ±yoruz.
# alpha argÃ¼manÄ± ile noktalarÄ± ÅŸeffaf hale getirebiiyoruz.
plt.scatter(x, y, s=sizes,alpha=0.5)

plt.show()
```

```python
# sutunlarÄ±n x deÄŸereri.
x = np.array(["A", "B", "C", "D"])
# sutunlarÄ±n y deÄŸereri.
y = np.array([3, 8, 1, 10])

# bar fonksiyonu ile sutun grafiklerini Ã§izdirebiliyoruz parametreler (x,y,rengi,sutunlarÄ±n kalÄ±nlÄ±klarÄ±).
plt.bar(x,y,color = "red",width = 0.1)

plt.show()
```

```python
# pasta grafiÄŸinde x ekseninin pozitif yÃ¶nÃ¼nden baÅŸlar ve saat yÃ¶nÃ¼nÃ¼n tersinde haraket eder. 
y = np.array([35, 25, 25, 15])

# parÃ§alarÄ± isimlendirmek iÃ§in isimleri bir liste iÃ§ine alÄ±yoruz.
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# parÃ§alar arasÄ±ndaki boÅŸluklarÄ± ifade eden deÄŸerleri listeliyoruz.
myexplode = [0.2, 0, 0, 0]

# lables parametresine isimleriizin listesinin bulunduÄŸu mylables deÄŸiÅŸkenini atÄ±yoruz, startangle parametresi ile grafiÄŸimizi dÃ¶ndÃ¼rebiliyoruz suan saat yÃ¶nÃ¼nÃ¼n tersine 90 derece dÃ¶ndÃ¼rdÃ¼k.
plt.pie(y,labels = mylabels, startangle = 90, explode = myexplode,shadow = True)
plt.show() 
```
